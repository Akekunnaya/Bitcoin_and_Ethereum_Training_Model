```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
Ethereum = pd.read_csv(r'C:\Users\aksha\Desktop\Projects\Bitcoins and Ethereum training model\Ethereum Historical Data.csv')
print("Data import successful")
Ethereum.describe()
```

```python
Ethereum.info()
```

```python
EthHigh = Ethereum.iloc[:, :-1].values  
EthLow = Ethereum.iloc[:, 1].values
```

```python
train_a, test_a, train_b, test_b = train_test_split(EthHigh, EthLow, test_size=0.2, random_state=0) 
print('Training Complete')
```

```python
regre = LinearRegression()  
regre.fit(train_a,train_b) 
```

```python
print(test_a) # Testing data - In High
pred_b = regre.predict(test_a) # Predicting the Low
pred_b
```

```python
df = pd.DataFrame({'Actual': test_b, 'Predicted': pred_b})  
df 
```

```python
print('Mean  error: ',mean_absolute_error(test_b,pred_b))
```
